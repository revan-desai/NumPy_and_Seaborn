```python
import numpy as np

mat1=np.array([[2,5,8],[6,9,3]])

t1=np.transpose(mat1)

print(t1)
```

```python
import numpy as np

mat1=np.array([[2,5,8],[6,9,3]])

result3=np.matmul(mat1,mat1.transpose())    # when you write mat1.transpose(instead of np.transpose()) it will make changes to original array

print(result3)
```
